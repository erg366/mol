```python
import random
import textwrap
import math

fryeText = open('frye.txt').read()
fryeWords = fryeText.split()

wikiText = open('deathWiki.txt').read()
wikiWords = wikiText.split()

walterText = open('whatDeathMeansNow.txt').read()
walterWords = walterText.split()
    
```

ABOVE: importing and reading the text files and splitting them up into lists

```python
randomPoem = []

for i in range(20):
    lineLength = random.randrange(12)
    spaceNum = random.randrange(40)
    lineWords = random.sample(fryeWords, lineLength)
    
    randomPoem.append(' ' * spaceNum + ' '.join(lineWords))
    print(randomPoem[i])
```

ABOVE: creating a new poem with a random word and format selection from the original list

```python
garbledPoem = []
cent = [1, 9]
year = [3, 2]

for i in range(len(randomPoem)):
    #lineWords = randomPoem[i].split() (this was my mistake)
    modChar = []
    for c in randomPoem[i]:
        if c.isalnum():
            modChar.append(chr(ord(c) + pow(random.choice(cent),random.choice(year))))
        else:
            modChar.append(c)
    garbledPoem.append(''.join(modChar))
    print(garbledPoem[i])
```

ABOVE: ran into the error ORD() EXPECTED A CHARACTER, BUT STRING OF LENGTH 2 FOUND, which i didn't understand, because i thought i was iterating through the strings on a character by character basis. i later realized that, because i was splitting randomPoem into lineWords and trying to iterate through that, c actually became the respective item in the list, which was a full word and thus could not be read by ord(). i picked 19 and 32 for my math operations because the do not stand at my grave was published in 1932.

```python
words = []
randomWords = []

for i in range(len(wikiWords)):
    if wikiWords[i].isalnum():
        words.append(wikiWords[i].lower())
    
for i in range(len(walterWords)):
    if walterWords[i].isalnum():
        words.append(walterWords[i].lower())
    
for i in range(400):
    randomWords.append(random.choice(words))
    
print(''.join(randomWords))
```

```python
waveWords = []

for i in range(len(randomWords)):
    for c in randomWords[i]:
        waveWords.append(chr(1032 - math.floor(math.sin(ord(c)) * 90)))
    
print(textwrap.fill(''.join(waveWords), 30))
```

ABOVE: i was not able to access the sin and cos functions that i wanted to, and got the error NAME 'MATH' IS NOT DEFINED, until i imported math further up in the code. it was interesting to play with the way different numbers changed the visual representation of the cos function.

```python
wordBins = []
wordBinsMod = []

for i in range(len(randomWords)):
    for c in randomWords[i]:
        wordBins.append(bin(ord(c) << 2))

for i in range(len(wordBins)):
    for c in wordBins[i]:
        if c.isalnum():
            if c == "b" and random.randint(0,2) == 1:
                wordBinsMod.append("\n")
            wordBinsMod.append(chr(ord(c) + 1 + 9 + 3 + 2))
        else:
            wordBinsMod.append("///")

print(''.join(wordBinsMod))
```
